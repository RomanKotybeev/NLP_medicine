```python
from yargy.tokenizer import MorphTokenizer


TOKENIZER = MorphTokenizer()
l = 'гиперемии'
list(TOKENIZER(l))
```

```python
datapath = pathlib.Path(r'E:\STUDY\8PM9I\Semester2\UIRS_2')
#temp_file = pathlib.Path(datapath).joinpath('tmp.txt')
docs = (file for file in datapath.iterdir() if re.search('\d.txt', file.name))
```

```python
texts = []

for doc in docs:
    with open(doc, 'rt', encoding='utf-8') as f:
        texts.append(f.read())
```

```python
from ipymarkup import show_markup
from yargy import and_, Parser, not_

parser = Parser(FEATURES)

text = texts[0]
matches = list(parser.findall(text))
spans = [_.span for _ in matches]
show_markup(text, spans)
```
